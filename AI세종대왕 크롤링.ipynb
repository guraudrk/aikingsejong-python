### 나무위키 세종대왕 문서 크롤링하기

html.parser는 BeautifulSoup 라이브러리에서 사용되는 html 파서이다.
파서는 html 문서를 파싱하고 구문 분석해서 파이썬이 이해할 수 있는 구조로 변환한다. 


```python
#BeautifulSoup을 통해 제어하자.

#url에서 html 가져오기
url ="https://namu.wiki/w/%EC%84%B8%EC%A2%85(%EC%A1%B0%EC%84%A0)"
response = requests.get(url)
response.encoding = 'utf-8' #utf-8로 특정화한다.
html_content = response.text

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_content, "html.parser")

# 텍스트 추출
text = soup.get_text()

# 결과 출력
print(text)

```

```python
#특정 class에서만 데이터를 추출함으로서 정보를 정재한다.

#클래스가 "Yt2xuBM2"인 요소들을 추출한다.
elements_with_class = soup.find_all(class_="Yt2xuBM2")

#이제 텍스트들을 추출한다.

texts = [element.get_text() for element in elements_with_class]



#결과 출력 

for text in texts:
    print(text)
```

```python
#이제 크롤링 한 정보에서 쓸데없는 데이터들은 모두 삭제한다.(숫자,기호 등)

import re #파이썬에서 정규 표현식을 사용할 때, re을 쓴다.


#한글만 남기고 나머지를 제거하는 함수

def remove_non_korean(text):
    korean_text = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣\s]', '', text)
    return korean_text

#텍스트에서 한글만 남기고 나머지를 제거한다.

cleaned_text = remove_non_korean(text)

#결과를 출력한다.
print(cleaned_text)
```
